```python
num_adults_tr = len(os.listdir(train_adults_dir))
num_children_tr = len(os.listdir(train_children_dir))
num_adults_tes = len(os.listdir(test_adults_dir))
num_children_tes = len(os.listdir(test_children_dir))
print('Adults Training data ',num_adults_tr)
print('Dog Training Data ',num_children_tr)
print('Adults Testing data ',num_adults_tes)
print('Children Testing data ',num_children_tes)
```

```python
total_train=num_adults_tr+num_children_tr
total_test=num_adults_tes+num_children_tes
print('Total Training images = ',total_train)
print('Total Testing images = ',total_test)
```

```python
batch_size = 70
epochs = 25
IMG_HIEGHT = 150
IMG_WIDHT = 150
```

```python
train_image_generator = ImageDataGenerator(rescale=1./255)
test_image_generator = ImageDataGenerator(rescale=1./255)
```

```python
train_data_gen = train_image_generator.flow_from_directory(batch_size=batch_size,
                                                           directory=train_dir,
                                                           shuffle=True,
                                                           target_size=(IMG_HIEGHT,IMG_WIDHT),
                                                           class_mode='binary')
```

```python
train_data_gen.class_indices
```

```python
tes_data_gen = test_image_generator.flow_from_directory(batch_size=batch_size,
                                                              directory=test_dir,
                                                              target_size=(IMG_HIEGHT,IMG_WIDHT),
                                                              class_mode='binary')
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1,5,figsize=(20,20))
    axes = axes.flatten()
    for img,ax in zip(images_arr,axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(images[10:15])
```

```python
model = Sequential([
    Conv2D(16,3,padding='same',activation='relu',input_shape=(IMG_HIEGHT,IMG_WIDHT,3)), #150
    MaxPooling2D(), #75
    Conv2D(32,3,padding='same',activation='relu'), #75
    MaxPooling2D(),#37
    Conv2D(64,3,padding='same',activation='relu'),#37
    MaxPooling2D(),#18
    Flatten(),
    Dense(256,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(train_data_gen,epochs=epochs,
                             validation_data=tes_data_gen)
```

```python
history.history
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(epochs)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(epochs_range,acc,label='Training Accuracy')
plt.plot(epochs_range,val_acc,label='Val Accuracy')
plt.legend(loc=0)
plt.title('Accuracy')
plt.subplot(1,2,2)
plt.plot(epochs_range,loss,label='Training Loss')
plt.plot(epochs_range,val_loss,label='Val LOss')
plt.legend(loc=0)
plt.title('Loss')
plt.show()
```

```python
model_new = Sequential([
    Conv2D(16,3,padding='same',activation='relu',input_shape=(IMG_HIEGHT,IMG_WIDHT,3)), #150
    MaxPooling2D(), #75
    Dropout(0.2),
    Conv2D(32,3,padding='same',activation='relu'), #75
    MaxPooling2D(),#37
    Dropout(0.2),
    Conv2D(64,3,padding='same',activation='relu'),#37
    MaxPooling2D(),#18
    Dropout(0.2),
    Flatten(),
    Dense(512,activation='relu'),
    Dropout(0.2),
    Dense(1,activation='sigmoid')
])
```

```python
model_new.summary()
```

```python
epochs = 100
```

```python
model_new.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model_new.fit(train_data_gen,steps_per_epoch=total_train//batch_size,
                                 epochs=epochs,
                                 validation_data=tes_data_gen,
                                 validation_steps=total_test//batch_size)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(epochs)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(epochs_range,acc,label='Training Accuracy')
plt.plot(epochs_range,val_acc,label='Val Accuracy')
plt.legend(loc='lower right')
plt.title('Accuracy')
plt.subplot(1,2,2)
plt.plot(epochs_range,loss,label='Training Loss')
plt.plot(epochs_range,val_loss,label='Val LOss')
plt.legend(loc='upper right')
plt.title('Loss')
plt.show()
```
